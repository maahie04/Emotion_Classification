```python
import os
dataset_folder = "NOT BAD"
genders = []
ages = []

for filename in os.listdir(dataset_folder):
    if filename.endswith(".wav"):
        parts = filename.split("_")
        gender = parts[4]
        genders.append(gender)
        age = int(parts[5].split(".")[0])
        ages.append(age)
```

```python
from collections import Counter
gender_counts = Counter(genders)
```

```python
#Pie chart
import matplotlib.pyplot as plt

labels = gender_counts.keys()
sizes = gender_counts.values()
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)

plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Gender Distribution of Participants")
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(8, 6))
counts, bins, patches = plt.hist(ages, bins=20, color='skyblue', edgecolor='black')

# Create a formatter that divides displayed values by 3
def divide_by_3(x, pos):
    return f'{x/3:.0f}'

plt.gca().yaxis.set_major_formatter(FuncFormatter(divide_by_3))
plt.title("Age Distribution of Participants")
plt.xlabel("Age")
plt.ylabel("Number of Participants")
plt.grid(axis='y', alpha=0.75)
plt.show()
```
